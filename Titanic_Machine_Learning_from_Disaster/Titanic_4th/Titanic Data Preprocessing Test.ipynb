```python
# Pandas를 사용하여 Titanic raw data 불러오기
import pandas as pd

# train 변수에 불러온 csv 파일을 index_col을 지정해 저장합니다
train = pd.read_csv("data/train.csv", index_col="PassengerId")

# train 변수에 할당된 데이터 행렬의 크기를 확인
# 저장된 데이터의 행렬수를 확인해서, 내가 변경하고 튜닝한 데이터가 제대로 맞게 들어갔는지 확인합니다.
print(train.shape)

# head()를 사용하면 train 데이터의 상위 5개를 볼 수 있습니다.
train.head()
```

```python
# test 데이터도 같은 방식으로 진행합니다.
test = pd.read_csv("data/test.csv", index_col="PassengerId")

# 제대로 test 데이터가 할당되어 있는지 확인합니다.
print(test.shape)

# 상위 5개 출력하여 test 데이터의 형태를 살펴봅니다.
test.head()
```
